# SVM Example

```python
# Import Libraries

import numpy as np
import pandas as pd 
```

```python
adsDataset = pd.read_csv("SVM Data/ad.data", header = None, sep = ',', low_memory=False)
```

```python
adsDataset.head(12)

    # Note the last column is our outcome / answer
```

```python
# Check whether a given value is a missing value, if yes change it to NaN
def toNum(cell):
    try:
        return np.float(cell)
    except:
        return np.nan
```

```python
# Apply missing value check to a column / Pandas series
def seriestoNum(series):
    return series.apply(toNum)
```

```python
# Select the training data using iloc to grab all cell rows & columns, then replace ? with NaN

train_data=adsDataset.iloc[0:,0:-1].apply(seriestoNum)

    # Only go up to the last column, ie remove the 'answer' from the training set
```

```python
train_data.head(12)
```

```python
# Drop Nulls

train_data=train_data.dropna()

train_data.head(20)
```

```python
# Create function to convert ad. to 1, if not 0
def toLabel(str):
    if str=="ad.":
        return 1
    else:
        return 0
    
train_labels=adsDataset.iloc[train_data.index,-1]

train_labels.shape
```

```python
train_labels = train_labels.apply(toLabel)
```

```python
# Training Phase 
from sklearn.svm import LinearSVC 

clf = LinearSVC()

clf.fit(train_data[100:2300],train_labels[100:2300])
```

```python
# Test Phase
clf.predict(train_data.iloc[14].values.reshape(1,-1))
```
